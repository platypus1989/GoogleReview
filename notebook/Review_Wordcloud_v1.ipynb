```python
%matplotlib inline
import nltk
import numpy as np
import pandas as pd
import seaborn as sns
from __future__ import division
import os
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
os.listdir('../data')
```

```python
review_data = pd.read_csv('../data/review_data_v1.csv')
```

```python
review_data.head()
```

```python
review_data['review_text'] = review_data['review_text'].astype(str)
text = ' '.join(review_data['review_text'])
tokens = nltk.word_tokenize(text.decode('utf-8'))
```

```python
icecream_mask = np.array(Image.open(("../images/icecream.png")))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=icecream_mask,
               stopwords=stopwords)
```

```python
# generate word cloud
wc.generate(text)
```

```python
wc.to_file(("../images/icecream_wordcloud.png"))
```

```python
# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(icecream_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
